```python
from smolagents import CodeAgent, DuckDuckGoSearchTool, InferenceClientModel
```

```python
model = InferenceClientModel()

agent = CodeAgent(
    tools=[DuckDuckGoSearchTool()],
    model=model
)
```

```python
result = agent.run(
    "Search for the best music recommendations for a superhero party."
)

print(result)
```
