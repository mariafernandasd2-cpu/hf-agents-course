```python
from typing_extensions import TypedDict
from langgraph.graph import StateGraph, START, END
```

```python
class EmailState(TypedDict):
    email: str
    category: str
```

```python
def read_email(state):
    
    print("Leyendo email...")
    
    return {}
```

```python
def classify_email(state):
    
    email = state["email"].lower()
    
    if "win money" in email:
        category = "spam"
    else:
        category = "legitimate"
    
    return {"category": category}
```

```python
def handle_spam(state):
    
    print("Email marcado como spam")
    
    return {}
```

```python
def handle_legitimate(state):
    
    print("Email legítimo")
    
    return {}
```

```python
def route_email(state):
    
    return state["category"]
```

```python
builder = StateGraph(EmailState)

builder.add_node("read_email", read_email)
builder.add_node("classify_email", classify_email)
builder.add_node("handle_spam", handle_spam)
builder.add_node("handle_legitimate", handle_legitimate)

builder.add_edge(START, "read_email")
builder.add_edge("read_email", "classify_email")

builder.add_conditional_edges(
    "classify_email",
    route_email,
    {
        "spam": "handle_spam",
        "legitimate": "handle_legitimate"
    }
)

builder.add_edge("handle_spam", END)
builder.add_edge("handle_legitimate", END)

graph = builder.compile()
```

```python
result = graph.invoke({
    "email": "Win money now",
    "category": ""
})

print(result)
```
