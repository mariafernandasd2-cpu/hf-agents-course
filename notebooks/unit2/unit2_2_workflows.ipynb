```python
from llama_index.core.workflow import (
    StartEvent,
    StopEvent,
    Workflow,
    step
)
```

```python
class MyWorkflow(Workflow):

    @step
    async def my_step(self, ev: StartEvent) -> StopEvent:

        return StopEvent(
            result="Workflow completed successfully!"
        )
```

```python
w = MyWorkflow(timeout=10, verbose=False)
```

```python
result = await w.run()

print(result)
```
