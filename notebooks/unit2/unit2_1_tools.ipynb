```python
from smolagents import tool

@tool
def catering_service_tool(query: str) -> str:
    """
    Returns the best catering service.

    Args:
        query: Search query.
    """

    services = {
        "Gotham Catering": 4.9,
        "Wayne Catering": 4.8,
        "Superhero Events": 4.7
    }

    return max(services, key=services.get)
```

```python
print(
    catering_service_tool(
        "Best catering service"
    )
)
```
